```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import torch
import PIL
import torchvision.transforms as transforms
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,multilabel_confusion_matrix
import cv2
from tensorflow import keras
from keras_tuner.engine.hyperparameters import HyperParameters
from keras import layers, models, regularizers
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import seaborn as sns
from sklearn.semi_supervised import LabelPropagation
from torchstat import stat
from sklearn.tree import DecisionTreeClassifier
import joblib
```

### Load all the images from the device

```python
file = []
labels = []
resized_image = []
for dirname,_, filenames in os.walk('lung_colon_image_set'):
    for filename in filenames:
        if not filename.lower().endswith('.ds_store'):
            if "lung_aca" in dirname:
                labels.append("Lung Adenocarcinoma")
            elif "lung_n" in dirname:
                labels.append("Lung Benign Tissue")
            elif "lung_scc" in dirname:
                labels.append("Lung Squamous Cell Carcinoma")
            elif "colon_n" in dirname:
                labels.append("Colon Benign Tissue")
            elif "colon_aca" in dirname:
                labels.append("Colon Adenocarcinoma")
            file.append(os.path.join(dirname, filename))
```

### Convert image path and labels into dataframe

```python
Slabels = pd.Series(labels,name="Labels")
Sfile  = pd.Series(file,name="Filepaths")
data = pd.concat([Sfile,Slabels],axis=1)
```

```python
data.head()
```

```python
traindf,validationdf = train_test_split(data,random_state=25,train_size=0.8,stratify = data['Labels'])
validationdf,testdf = train_test_split(validationdf,random_state=25,train_size=0.5 ,stratify = validationdf['Labels'])
```

### Create three diffrerent generator with multiple preprocessing on images
Here, we have done resizing the images from 768*768 to 64*64 and the dataset is splitted with batch size of 10.

```python
# Set the batch size and image dimensions
image_size = (64, 64)  # Adjust this according to your image size

# Create the ImageDataGenerator objects
datagen = ImageDataGenerator(rescale=1.0 / 255)
train_generator = datagen.flow_from_dataframe(
    traindf,
    x_col='Filepaths',
    y_col='Labels',
    target_size=(64, 64),
    batch_size=10,  # Adjust as needed
    class_mode='categorical',  # Adjust based on your label format
    shuffle=False,
    seed=42
)

validation_generator = datagen.flow_from_dataframe(
    validationdf,
    x_col='Filepaths',
    y_col='Labels',
    target_size=(64, 64),
    batch_size=10,  # Adjust as needed
    class_mode='categorical',  # Adjust based on your label format
    shuffle=False,
    seed=42
)

test_generator = datagen.flow_from_dataframe(
    testdf,
    x_col='Filepaths',
    y_col='Labels',
    target_size=(64, 64),
    batch_size=10,  # Adjust as needed
    class_mode='categorical',  # Adjust based on your label format
    shuffle=False,
    seed=42
)
```

### Convert images pixels into numpy array and flatten the array 

```python
def get_data_from_generator(generator):
    num_samples = generator.samples

    # Initialize empty arrays to store the features and labels
    X = []
    y = []

    # Iterate over all the batches and collect the features and labels
    for _ in range(num_samples // generator.batch_size):
        batch_X, batch_y = generator.next()
        X.extend(batch_X)
        y.extend(batch_y)

    # Convert the collected data to numpy arrays
    X = np.array(X)
    y = np.argmax(y,axis=1)

    # Reshape the image data if needed
    if len(X.shape) > 2:
        X = X.reshape(-1, *X.shape[1:])

    return X.reshape(X.shape[0], -1), y
```

```python
print("TRAIN:")
trainX,trainY = get_data_from_generator(train_generator)
print(len(trainX),len(trainY))

print("VALIDATION:")
validationX,validationY = get_data_from_generator(validation_generator)
print(len(validationX),len(validationY))

print("TEST:")
testX,testY = get_data_from_generator(test_generator)
print(len(testX),len(testY))
print(len(trainX[0]))
```

 In the context of semi-supervised learning, we have partitioned the training dataset into labeled and unlabeled data, maintaining a ratio of 1:3 between them.

```python
labeled_train_df,unlabeled_train_df = train_test_split(traindf,train_size=0.25,stratify
                                                       = traindf["Labels"])
```

```python
labeled_train_generator = datagen.flow_from_dataframe(
    labeled_train_df,
    x_col='Filepaths',
    y_col='Labels',
    target_size=(64, 64),
    batch_size=10,  # Adjust as needed
    class_mode='categorical',  # Adjust based on your label format
    shuffle=False,
    seed=42
)

unlabeled_train_generator = datagen.flow_from_dataframe(
    unlabeled_train_df,
    x_col='Filepaths',
    y_col='Labels',
    target_size=(64, 64),
    batch_size=10,  # Adjust as needed
    class_mode='categorical',  # Adjust based on your label format
    shuffle=False,
    seed=42
)
```

```python
print("TRAIN LABELED DATASET:")
labeled_trainX,labeled_trainY = get_data_from_generator(labeled_train_generator)
print(len(labeled_trainX),len(labeled_trainY))
```

```python
print("TRAIN UNLABELED DATASET:")
unlabeled_trainX,unlabeled_trainY = get_data_from_generator(unlabeled_train_generator)
print(len(unlabeled_trainX),len(unlabeled_trainY))
```

```python
classifier = DecisionTreeClassifier(max_depth=10,criterion='entropy',min_samples_split=100)
classifier.fit(labeled_trainX,labeled_trainY)
```

**getSelectedImageIndexes** will provide the list of index which have more than 0.75% of predication probability.

```python
def getSelectedImageIndexes(probs):
    selected_images = []
    un_selected_images = []
    for index in range(len(probs)):
        if max(probs[index]) >= 0.75:
            selected_images.append(index)
        else:
            un_selected_images.append(index)
    return [selected_images,un_selected_images]
```

**getScoreCard** is used to show the details of the current model iteration

```python
test_accuracy_list = []
train_accuracy_list = []
validation_accuracy_list = []
def getScoreCard():
    y_val_pred1 = classifier.predict(validationX)
    val_accuracy1 = accuracy_score(validationY, y_val_pred1)

    # Evaluate the model on the test set
    y_test_pred1 = classifier.predict(testX)
    test_accuracy1 = accuracy_score(testY, y_test_pred1)

    train_accuracy1 = accuracy_score(labeled_trainY,classifier.predict(labeled_trainX))
    overfitting = round(((train_accuracy1-test_accuracy1)/train_accuracy1)*100,2)
    numbers = [val_accuracy1,test_accuracy1]
    labels = ["Validation","Test"]
    train_accuracy_list.append(train_accuracy1)
    test_accuracy_list.append(test_accuracy1)
    validation_accuracy_list.append(val_accuracy1)
    print("-----------------------------")
    print('| {:>10s} | {:>10s} |'.format(*labels))
    print('| {:>10.4f} | {:>10.4f} |'.format(*numbers))
    print("-------------------------------")
```

### Model training

```python
%%time
iteration = 0

while unlabeled_trainX.shape[0] > 0:
    iteration += 1
    print("Iteration::" + str(iteration) + "trainX:" + str(len(unlabeled_trainX)))
    if unlabeled_trainX.shape[0] <= 0:
        break
    classifier.fit(labeled_trainX,labeled_trainY)
    unlabeled_train_pridictions = classifier.predict(unlabeled_trainX)
    unlabeled_train_probs = classifier.predict_proba(unlabeled_trainX)
    filtered_list = getSelectedImageIndexes(unlabeled_train_probs)
    selected_image_indexes = filtered_list[0]
    if(len(selected_image_indexes)) == 0:
        break;
    un_selected_image_indexes = filtered_list[1]
    print("selected_image_indexes:"+str(len(selected_image_indexes)))
    print("un_selected_image_indexes:"+str(len(un_selected_image_indexes)))
    if len(un_selected_image_indexes) == unlabeled_trainX.shape[0]:
        break
        
    labeled_trainX = np.concatenate((labeled_trainX,np.array(unlabeled_trainX)[selected_image_indexes,:]))
    labeled_trainY = np.concatenate((labeled_trainY,np.array(unlabeled_train_pridictions)[selected_image_indexes]))
#     for selected_index in selected_image_indexes:
#         labeled_trainX = np.concatenate((labeled_trainX,[unlabeled_trainX[selected_index]]))
#         labeled_trainY = np.concatenate((labeled_trainY,[unlabeled_train_pridictions[selected_index]]))
    unlabeled_trainX = np.delete(unlabeled_trainX,selected_image_indexes,axis=0)   
    getScoreCard()
```

```python
y_val_pred = classifier.predict(validationX)
val_accuracy = accuracy_score(validationY, y_val_pred)

# Evaluate the model on the test set
y_test_pred = classifier.predict(testX)
test_accuracy = accuracy_score(testY, y_test_pred)

print("Validation Accuracy:", val_accuracy)
print("Test Accuracy:", test_accuracy)
print("Train Accuracy:",accuracy_score(trainY,classifier.predict(trainX)))
```
